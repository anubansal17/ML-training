```python
X= quality.iloc[:,0:13] #it excludes 13th wala (leaving poor care that is dependent variables)( hence X has independent columns)
```

```python
y=quality["PoorCare"]
```

```python
#jb hum linear regression kr re the tb jo model we didn't needed to specify x and y
```

```python
#in logistic regression we have specify x and y
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=88,stratify=y)
```

```python
#stratify se it maintains the same ratio of poor and good care as in train data
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1= LogisticRegression()
```

```python
model1.fit(X_train[["Narcotics","OfficeVisits"]],y_train)
```

```python
model1.coef_
```

```python
model1.score(X_train[["Narcotics","OfficeVisits"]],y_train)
```

```python
model1.score(X_test[["Narcotics","OfficeVisits"]],y_test)
```

```python
pred_train=model1.predict(X_train[["Narcotics","OfficeVisits"]])
```
